```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import models,layers
from sklearn.model_selection import train_test_split
from tensorflow.keras.regularizers import l2
```

```python
data=pd.read_csv('creditcard (1).csv')
data
```

```python
data.isnull().sum()
```

```python
data
```

```python
time_means=data.Time.mean(axis=0)
time_std=data.Time.std(axis=0)
data.Time-=time_means
data.Time/=time_std
```

```python
amount=data.Amount.mean(axis=0)
amount_std=data.Amount.std(axis=0)
data.Amount-=amount
data.Amount/=amount_std
data
```

```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1]
```

```python
x_tran,x_test,y_tran,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
x_tran, x_val, y_tran, y_val = train_test_split(x_tran, y_tran, test_size=0.3, random_state=0)
```

```python
model=models.Sequential()
model.add(layers.Dense(6, activation= 'relu',kernel_regularizer=l2(0.01),input_shape=(30,)))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(6, activation= 'relu',kernel_regularizer=l2(0.01)))
model.add(layers.Dense(1,activation='sigmoid'))
model.compile(optimizer='adam',loss='binary_crossentropy', metrics= ['accuracy'])
model.fit(x_tran, y_tran,epochs=5,validation_data=(x_val, y_val))
```

```python
model.evaluate(x_test,y_test)
```
